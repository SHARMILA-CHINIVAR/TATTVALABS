```python
import matplotlib
from matplotlib import pyplot as pl,style
from mpl_toolkits.mplot3d import Axes3D
style.use('grayscale')
import pandas as pd
import csv
fig=pl.figure()
ax=fig.add_subplot(111,projection='3d')
df=pd.read_csv(r'F:\TATTVALABS\iris.csv')
n=len(df)
for i in range(0,n):
    if str(df.iloc[i]['species'])=='setosa':
        ax.scatter(df.iloc[i]['sepal_length'],df.iloc[i]['sepal_width'],color='r')
    if str(df.iloc[i]['species'])=='versicolor':
        ax.scatter(df.iloc[i]['sepal_length'],df.iloc[i]['sepal_width'],color='b')
    if str(df.iloc[i]['species'])=='virginica':
        ax.scatter(df.iloc[i]['sepal_length'], df.iloc[i]['sepal_width'], color='g')
ax.set_xlabel('Sepal_Length')
ax.set_ylabel('Sepal_Width')
pl.savefig('F:\\TATTVALABS\\scatter_pic.png')
pl.show()
```

```python
import matplotlib
from matplotlib import pyplot as pl
import pandas as pd
import csv
df=pd.read_csv(r'F:\TATTVALABS\iris.csv')
pl.figure(figsize = (10, 7)) 
x = df["sepal_length"] 
pl.hist(x, bins = 20, color = "green") 
pl.title("SEPAL LENGTH") 
pl.xlabel("Sepal_Length_cm") 
pl.ylabel("Count") 
pl.show()
```
